```python
#import dependencies 
import pandas as pd 
```

```python
#load in csv and preview 
degrees_df = pd.read_csv('resources/degrees-that-pay-back.csv')
degrees_df.head()
```

```python
#rename columns 
degrees_df = degrees_df.rename(columns={"Undergraduate Major": "undergraduate_major", "Starting Median Salary": "starting_median_salary","Mid-Career Median Salary":"mid_career_median_salary", "Percent change from Starting to Mid-Career Salary":"percent_change_from_starting_to_mid_career_salary","Mid-Career 10th Percentile Salary":"mid_career_10th_percentile_salary","Mid-Career 25th Percentile Salary":"mid_career_25th_percentile_salary","Mid-Career 75th Percentile Salary":"mid_career_75th_percentile_salary","Mid-Career 90th Percentile Salary":"mid_career_90th_percentile_salary"})
degrees_df.head()
```

```python
degrees_df['starting_median_salary'] = degrees_df['starting_median_salary'].str.replace('$', '').str.replace(',', '')
degrees_df['mid_career_median_salary'] = degrees_df['mid_career_median_salary'].str.replace('$', '').str.replace(',', '')
degrees_df['mid_career_10th_percentile_salary'] = degrees_df['mid_career_10th_percentile_salary'].str.replace('$', '').str.replace(',', '')
degrees_df['mid_career_25th_percentile_salary'] = degrees_df['mid_career_25th_percentile_salary'].str.replace('$', '').str.replace(',', '')
degrees_df['mid_career_75th_percentile_salary'] = degrees_df['mid_career_75th_percentile_salary'].str.replace('$', '').str.replace(',', '')
degrees_df['mid_career_90th_percentile_salary'] = degrees_df['mid_career_90th_percentile_salary'].str.replace('$', '').str.replace(',', '')
degrees_df.head()
```

```python
#read df back into csv 
degrees_df.to_csv("resources/degrees_that_pay_back_cleaned.csv",index = False,index_label = False)
```
